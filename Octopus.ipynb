```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from octopus.gd   import experiment_gd
from octopus.pgd  import experiment_pgd
from octopus.agd  import experiment_agd
from octopus.pagd import experiment_pagd
```

```python
iters = 1000
x     = [0 + 0.1 for i in range(15)]
```

```python
np.random.seed(10)

gd_vals   = experiment_gd(x, iters)
pgd_vals  = experiment_pgd(x, iters)
agd_vals  = experiment_agd(x, iters)
pagd_vals = experiment_pagd(x, iters)
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(8,8))

plt.plot(gd_vals,   label  = 'GD')
plt.plot(pgd_vals,  label  = 'PGD')
plt.plot(agd_vals,  label  = 'AGD')
plt.plot(pagd_vals, label  = 'PAGD')

plt.xlabel('Iterations')
plt.ylabel('$f(x_k)$')
plt.legend()
plt.savefig('figures/octopus.pdf', bbox_inches='tight')
plt.show()
```
